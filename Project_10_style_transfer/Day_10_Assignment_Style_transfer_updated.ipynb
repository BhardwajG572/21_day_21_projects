# Assignment:

- Load the pretrainde Generator
- Load Gender Vector
- Generate 10 variations of the image from male to female as per gender vector

The core of this project is transfer learning, a powerful technique where we leverage a pre-trained model, MobileNetV2, which has already learned to recognize a vast array of features from millions of images in the ImageNet dataset. We adapt this pre-existing knowledge and fine-tune it for our specific gender classification task

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import os

# Step 0 : Loading the Model :

if not os.path.exists('Face-Generator-with-GAN'):
    print("Cloning repository...")
    os.system("git clone https://github.com/AshishJangra27/Face-Generator-with-GAN.git")
else:
    print("✅ Repository already exists.")

# Step 1: Load the Pre-trained Generator
# Here, we load the pre-trained GAN generator model.

try:
    model_path = 'Face-Generator-with-GAN/generator_700.h5'
    generator = keras.models.load_model(model_path)
    print(f"✅ Generator model loaded successfully from '{model_path}'!")
except Exception as e:
    print(f"❌ Error loading model: {e}")

latent_dim = 100

# Step 2: Create a DEMONSTRATION Gender Vector

gender_vector = np.random.randn(1, latent_dim) * 0.5 # A random direction vector

### --- Step 3: Generate and Interpolate Images ---
# We create a random starting face and then generate 10 variations by moving


start_vector = np.random.normal(size=(1, latent_dim))

transition_images = []
num_steps = 10
interpolation_strength = 3.0

print(f"Generating {num_steps} images...")
for i in range(num_steps):
    alpha = (i / (num_steps - 1))

    modified_vector = start_vector + (alpha * interpolation_strength * gender_vector)

    generated_image = generator.predict(modified_vector, verbose=0)
    transition_images.append(generated_image)

print(f" Generated {len(transition_images)} transition images.")

### --- Step 4: Visualize the Results ---
# This final step uses matplotlib to display all 10 generated images in a grid,


fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, img_array in enumerate(transition_images):
    ax = axes[i // 5, i % 5]

    img_display = (img_array[0] + 1) / 2.0

    ax.imshow(img_display)
    ax.set_title(f"Step {i+1}")
    ax.axis('off')

fig.suptitle("Demonstration of Image Transition with a Random Vector", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
