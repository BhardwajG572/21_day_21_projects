# Assignment : Image Enhancer from (64x64 to 256x256) using U-NET

We first upscale the 64x64 image to an intermediate 128x128 resolution. This allows us to assess the quality of the enhancement before proceeding to the final 256x256 target.

```python
# Step 1: Import All Necessary Libraries
import os
import cv2
import numpy as np
import kagglehub
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Importing the library from kaggle hub
print("Downloading the CelebA dataset via KaggleHub...")
dataset_path = kagglehub.dataset_download("jessicali9530/celeba-dataset")
print(f"Dataset downloaded to: {dataset_path}")


IMAGE_DIRECTORY = os.path.join(dataset_path, 'img_align_celeba/img_align_celeba')
print(f"Image directory is set to: {IMAGE_DIRECTORY}")


# 3. Defining UNET Model Architecture
def unet_64to128(input_shape=(64, 64, 3)):
    """
    Defines the UNET model architecture.
    Takes a 64x64 image and outputs a 128x128 image.
    """
    inputs = Input(input_shape)

    # Encoder (Contracting Path)
    c1 = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    p1 = MaxPooling2D((2, 2))(c1) # 32x32

    c2 = Conv2D(64, (3, 3), activation='relu', padding='same')(p1)
    p2 = MaxPooling2D((2, 2))(c2) # 16x16

    c3 = Conv2D(128, (3, 3), activation='relu', padding='same')(p2)
    p3 = MaxPooling2D((2, 2))(c3) # 8x8

    c4 = Conv2D(256, (3, 3), activation='relu', padding='same')(p3)
    p4 = MaxPooling2D((2, 2))(c4) # 4x4

    # Bottleneck
    b = Conv2D(512, (3, 3), activation='relu', padding='same')(p4)
    b = Conv2D(512, (3, 3), activation='relu', padding='same')(b)

    # Decoder (Expansive Path)
    u1 = Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(b) # 8x8
    u1 = concatenate([u1, c4]) # Skip Connection
    c5 = Conv2D(256, (3, 3), activation='relu', padding='same')(u1)

    u2 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5) # 16x16
    u2 = concatenate([u2, c3]) # Skip Connection
    c6 = Conv2D(128, (3, 3), activation='relu', padding='same')(u2)

    u3 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6) # 32x32
    u3 = concatenate([u3, c2]) # Skip Connection
    c7 = Conv2D(64, (3, 3), activation='relu', padding='same')(u3)

    u4 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c7) # 64x64
    u4 = concatenate([u4, c1]) # Skip Connection
    c8 = Conv2D(32, (3, 3), activation='relu', padding='same')(u4)

    # Final Upsampling to 128x128
    u5 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c8) # 128x128
    c9 = Conv2D(16, (3, 3), activation='relu', padding='same')(u5)

    outputs = Conv2D(3, (1, 1), activation='sigmoid')(c9) # Output layer

    model = Model(inputs=[inputs], outputs=[outputs])
    return model

# 4: Create the Data Generator
def datagen(image_dir, batch_size=16):
    """
    A generator to feed image data to the model in batches.
    """
    image_files = os.listdir(image_dir)

    while True:
        batch_files = np.random.choice(image_files, size=batch_size)

        x_batch, y_batch = [], []

        for file in batch_files:
            try:
                img_path = os.path.join(image_dir, file)
                img = cv2.imread(img_path)
                if img is None: continue

                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img_128 = cv2.resize(img, (128, 128))
                img_64 = cv2.resize(img_128, (64, 64))

                x_batch.append(img_64 / 255.0)
                y_batch.append(img_128 / 255.0)
            except Exception as e:
                # print(f"Skipping file {file} due to error: {e}")
                continue

        yield (np.array(x_batch), np.array(y_batch))

# 5: Build and Compile the Model
print("\nBuilding and compiling the UNET model...")
model = unet_64to128()
model.compile(optimizer=Adam(learning_rate=0.001), loss='mae')
model.summary()

# 6: Train the Model
print("\nStarting model training...")
BATCH_SIZE = 32
STEPS_PER_EPOCH = 150
EPOCHS = 5

train_generator = datagen(IMAGE_DIRECTORY, batch_size=BATCH_SIZE)

history = model.fit(
    train_generator,
    steps_per_epoch=STEPS_PER_EPOCH,
    epochs=EPOCHS
)
print("Training complete.")

# Saving the trained model
model.save('unet_face_enhancement_colab.h5')
print("Model saved to unet_face_enhancement_colab.h5")

# 7: Test the Model and Visualize Results
def test_and_visualize(model, image_dir, num_samples=3):
    """
    Selects random images, enhances them, and displays the results.
    """
    print("\nVisualizing results with test images...")
    image_files = np.random.choice(os.listdir(image_dir), num_samples)

    plt.figure(figsize=(12, 4 * num_samples))

    for i, file in enumerate(image_files):
        img_path = os.path.join(image_dir, file)

        original_img = cv2.imread(img_path)
        original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

        img_128 = cv2.resize(original_img, (128, 128))
        img_64 = cv2.resize(img_128, (64, 64))

        input_img = np.expand_dims(img_64 / 255.0, axis=0)
        predicted_img_128 = model.predict(input_img)[0]

        plt.subplot(num_samples, 3, i * 3 + 1)
        plt.title("Low-Res Input (64x64)")
        plt.imshow(img_64)
        plt.axis('off')

        plt.subplot(num_samples, 3, i * 3 + 2)
        plt.title("Model Output (128x128)")
        plt.imshow(predicted_img_128)
        plt.axis('off')

        plt.subplot(num_samples, 3, i * 3 + 3)
        plt.title("Ground Truth (128x128)")
        plt.imshow(img_128)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

test_and_visualize(model, IMAGE_DIRECTORY)
```

## AI Face Resolution Enhancement with UNET

This project demonstrates an AI-powered image super-resolution technique to enhance low-resolution face images. It uses a UNET, a powerful deep learning architecture, to transform blurry 64x64 pixel images into sharp, detailed 256x256 pixel portraits


## Model Architecture
DataSet : we have used CelebA dataset , imported from KaggleHub
The core of this project is the U-NET model. Its U-shaped design is perfect for image-to-image tasks.

- Encoder (Contracting Path): Analyzes the input image to understand its high-level features (the "what").

- Bottleneck: Creates the most condensed summary of the image's content.

- Decoder (Expansive Path): Reconstructs the image from the summary, adding back details.

- Skip Connections: The key feature that feeds high-resolution details from the encoder directly to the decoder, preventing detail loss and ensuring the output is sharp.


## About U-NET 🇰
At its core, UNET is a powerful type of neural network architecture designed specifically for tasks where the output is also an image, like in our project. It's famous for being incredibly good at image segmentation (figuring out exactly which pixels in an image belong to an object) and image enhancement (like your task of increasing resolution).

The name "UNET" comes from the fact that its architecture diagram looks like the letter U.

Think of it like an artist who first squints to see the overall shape of a subject and then puts on their glasses to fill in the fine details. The UNET works in a similar, two-part process.

## 1. The Encoder (The "Down" Part of the U) 🎨
The first half of the network is the encoder or contracting path. Its job is to analyze and understand the content of the input image.

It takes our 64x64 low-resolution image and repeatedly processes it through several layers.

With each step, the image representation gets smaller (it's "down-sampled"), but the network's understanding of the features in the image gets deeper.

By the bottom of the "U," the network has a very compressed, high-level understanding of the image. It doesn't care about the exact pixel locations anymore; it just knows, "This is a face, the eyes are roughly here, and the nose is roughly there."

Between Encoder and Decoder we have BottleNeck
- **Bottleneck :** The bottleneck layer contains the most condensed summary of the image's content before the network starts the process of reconstructing the image in higher resolution (the decoder path). It's a crucial part of the network as it holds the core "understanding" of the image that the decoder will use to build the output.


## 2. The Decoder (The "Up" Part of the U) ✨
The second half of the network is the decoder or expansive path. Its job is to take that compressed understanding and rebuild the image in high resolution.

It starts with the compressed feature information from the bottom of the "U."

It then repeatedly "up-samples" this information, gradually increasing the size of the image representation.

As it reconstructs the image, it focuses on adding back the fine details and textures needed for the high-resolution output (your 256x256 image).

## The Secret Sauce: Skip Connections 🔗
So, what makes the "U" shape so special? It's the skip connections.

These are connections that go directly from the encoder side to the decoder side, "skipping" the bottom of the U.

###### **Why is this so important?**
When the encoder analyzes the image, it captures high-resolution feature details in the early layers (like sharp edges). When the decoder is rebuilding the image, it can lose some of this precise location information.

The skip connections feed that detailed, high-resolution information directly to the decoder at the right moments. This allows the decoder to combine the general context ("what this is") from the bottom of the U with the precise details ("where it is") from the skip connections.

In short, for our project: The UNET's encoder learns what a face looks like from our low-res image, and the decoder, with the help of skip connections, uses that knowledge to intelligently paint a brand new, detailed, high-resolution version of it.

```python
# 1-- Importing  all the necessary libraries
import os
import cv2
import numpy as np
import kagglehub
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

#2--- Download  the Dataset from KaggleHub
print("Downloading the CelebA dataset via KaggleHub...")
dataset_path = kagglehub.dataset_download("jessicali9530/celeba-dataset")
print(f"Dataset downloaded to: {dataset_path}")

IMAGE_DIRECTORY = os.path.join(dataset_path, 'img_align_celeba/img_align_celeba')
print(f"Image directory is set to: {IMAGE_DIRECTORY}")


# 3-- Define the MODIFIED UNET Model Architecture
def unet_64to256(input_shape=(64, 64, 3)):
    """
    Defines the UNET model architecture.
    Takes a 64x64 image and outputs a 256x256 image.

    (64x64 ----> 256x256 image)

    """
    inputs = Input(input_shape)

    c1 = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    p1 = MaxPooling2D((2, 2))(c1)
    c2 = Conv2D(64, (3, 3), activation='relu', padding='same')(p1)
    p2 = MaxPooling2D((2, 2))(c2)
    c3 = Conv2D(128, (3, 3), activation='relu', padding='same')(p2)
    p3 = MaxPooling2D((2, 2))(c3)
    c4 = Conv2D(256, (3, 3), activation='relu', padding='same')(p3)
    p4 = MaxPooling2D((2, 2))(c4)

   #-- Bottleneck

    b = Conv2D(512, (3, 3), activation='relu', padding='same')(p4)
    b = Conv2D(512, (3, 3), activation='relu', padding='same')(b)

    # --- Decoder (Expansive Path) --- (MODIFIED)
    u1 = Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(b)
    u1 = concatenate([u1, c4])
    c5 = Conv2D(256, (3, 3), activation='relu', padding='same')(u1)

    u2 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5)
    u2 = concatenate([u2, c3])
    c6 = Conv2D(128, (3, 3), activation='relu', padding='same')(u2)

    u3 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6)
    u3 = concatenate([u3, c2])
    c7 = Conv2D(64, (3, 3), activation='relu', padding='same')(u3)

    u4 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c7)
    u4 = concatenate([u4, c1])
    c8 = Conv2D(32, (3, 3), activation='relu', padding='same')(u4) # Output is 64x64 here

    # --- Final Upsampling Layers ---
    u5 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c8) # Upsamples to 128x128
    c9 = Conv2D(16, (3, 3), activation='relu', padding='same')(u5)

    # --- Block --- >  >>> 128x128 to 256x256 ---
    u6 = Conv2DTranspose(8, (2, 2), strides=(2, 2), padding='same')(c9) # Upsamples to 256x256
    c10 = Conv2D(8, (3, 3), activation='relu', padding='same')(u6)

    outputs = Conv2D(3, (1, 1), activation='sigmoid')(c10) # MODIFIED to use final layer c10

    model = Model(inputs=[inputs], outputs=[outputs])
    return model

# 4: MODIFIED Data Generator
def datagen(image_dir, batch_size=16):
    """
    A generator to feed image data to the model in batches.
    MODIFIED to create 64x64 inputs and 256x256 ground truth outputs.
    """
    image_files = os.listdir(image_dir)

    while True:
        batch_files = np.random.choice(image_files, size=batch_size)

        x_batch, y_batch = [], []

        for file in batch_files:
            try:
                img_path = os.path.join(image_dir, file)
                img = cv2.imread(img_path)
                if img is None: continue

                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # MODIFIED: Ground truth is now 256x256
                img_256 = cv2.resize(img, (256, 256))

                # MODIFIED: Low-res input is created from the 256x256 image
                img_64 = cv2.resize(img_256, (64, 64))

                x_batch.append(img_64 / 255.0)
                y_batch.append(img_256 / 255.0) # MODIFIED
            except Exception as e:
                continue

        yield (np.array(x_batch), np.array(y_batch))

# 5: Build and Compile the Model
print("\nBuilding and compiling the UNET 64->256 model...")
model = unet_64to256() # MODIFIED to call the new function
model.compile(optimizer=Adam(learning_rate=0.001), loss='mae')
model.summary()

# 6: Train the Model
print("\nStarting model training...")
BATCH_SIZE = 16 # Reduced batch size as 256x256 images use more memory
STEPS_PER_EPOCH = 150
EPOCHS = 5

train_generator = datagen(IMAGE_DIRECTORY, batch_size=BATCH_SIZE)

history = model.fit(
    train_generator,
    steps_per_epoch=STEPS_PER_EPOCH,
    epochs=EPOCHS
)
print("Training complete.")
model.save('unet_face_enhancement_256.h5')
print("Model saved to unet_face_enhancement_256.h5")

# 7: Test the Model and MODIFIED Visualization
def test_and_visualize(model, image_dir, num_samples=3):
    print("\nVisualizing results with test images...")
    image_files = np.random.choice(os.listdir(image_dir), num_samples)

    plt.figure(figsize=(12, 4 * num_samples))

    for i, file in enumerate(image_files):
        img_path = os.path.join(image_dir, file)

        original_img = cv2.imread(img_path)
        original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

        # MODIFIED: Ground truth is 256x256
        img_256 = cv2.resize(original_img, (256, 256))
        img_64 = cv2.resize(img_256, (64, 64))

        input_img = np.expand_dims(img_64 / 255.0, axis=0)
        predicted_img_256 = model.predict(input_img)[0] # MODIFIED variable name

        plt.subplot(num_samples, 3, i * 3 + 1)
        plt.title("Low-Res Input (64x64)")
        plt.imshow(img_64)
        plt.axis('off')

        plt.subplot(num_samples, 3, i * 3 + 2)
        plt.title("Model Output (256x256)") # MODIFIED
        plt.imshow(predicted_img_256) # MODIFIED
        plt.axis('off')

        plt.subplot(num_samples, 3, i * 3 + 3)
        plt.title("Ground Truth (256x256)") # MODIFIED
        plt.imshow(img_256) # MODIFIED
        plt.axis('off')

    plt.tight_layout()
    plt.show()

test_and_visualize(model, IMAGE_DIRECTORY)
```

```python
import os
import cv2
import numpy as np
import kagglehub
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

print("Downloading the CelebA dataset via KaggleHub...")
dataset_path = kagglehub.dataset_download("jessicali9530/celeba-dataset")
IMAGE_DIRECTORY = os.path.join(dataset_path, 'img_align_celeba/img_align_celeba')
print(f"Image directory is set to: {IMAGE_DIRECTORY}")

all_files = os.listdir(IMAGE_DIRECTORY)

# we are using smaller subset for a quicker demonstration

subset_files = all_files[:20000]
train_files, val_files = train_test_split(subset_files, test_size=0.1, random_state=42)
print(f"Training on {len(train_files)} images, validating on {len(val_files)} images.")

def unet_64to256(input_shape=(64, 64, 3)):
    inputs = Input(input_shape)
    c1 = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    p1 = MaxPooling2D((2, 2))(c1)
    c2 = Conv2D(64, (3, 3), activation='relu', padding='same')(p1)
    p2 = MaxPooling2D((2, 2))(c2)
    c3 = Conv2D(128, (3, 3), activation='relu', padding='same')(p2)
    p3 = MaxPooling2D((2, 2))(c3)
    c4 = Conv2D(256, (3, 3), activation='relu', padding='same')(p3)
    p4 = MaxPooling2D((2, 2))(c4)
    b = Conv2D(512, (3, 3), activation='relu', padding='same')(p4)
    b = Conv2D(512, (3, 3), activation='relu', padding='same')(b)
    u1 = Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(b)
    u1 = concatenate([u1, c4])
    c5 = Conv2D(256, (3, 3), activation='relu', padding='same')(u1)
    u2 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5)
    u2 = concatenate([u2, c3])
    c6 = Conv2D(128, (3, 3), activation='relu', padding='same')(u2)
    u3 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6)
    u3 = concatenate([u3, c2])
    c7 = Conv2D(64, (3, 3), activation='relu', padding='same')(u3)
    u4 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c7)
    u4 = concatenate([u4, c1])
    c8 = Conv2D(32, (3, 3), activation='relu', padding='same')(u4)
    u5 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c8)
    c9 = Conv2D(16, (3, 3), activation='relu', padding='same')(u5)
    u6 = Conv2DTranspose(8, (2, 2), strides=(2, 2), padding='same')(c9)
    c10 = Conv2D(8, (3, 3), activation='relu', padding='same')(u6)
    outputs = Conv2D(3, (1, 1), activation='sigmoid')(c10)
    model = Model(inputs=[inputs], outputs=[outputs])
    return model

def datagen(image_files, image_dir, batch_size=16):
    while True:
        batch_files = np.random.choice(image_files, size=batch_size)
        x_batch, y_batch = [], []
        for file in batch_files:
            try:
                img_path = os.path.join(image_dir, file)
                img = cv2.imread(img_path)
                if img is None: continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img_256 = cv2.resize(img, (256, 256))
                img_64 = cv2.resize(img_256, (64, 64))
                x_batch.append(img_64 / 255.0)
                y_batch.append(img_256 / 255.0)
            except Exception as e:
                continue
        yield (np.array(x_batch), np.array(y_batch))

print("\nBuilding and compiling the UNET 64->256 model...")
model = unet_64to256()
model.compile(optimizer=Adam(learning_rate=0.001), loss='mae')
model.summary()

print("\nStarting model training...")
BATCH_SIZE = 16
STEPS_PER_EPOCH = len(train_files) // BATCH_SIZE
VALIDATION_STEPS = len(val_files) // BATCH_SIZE
EPOCHS = 10      # we are using higher number of epochs

train_generator = datagen(train_files, IMAGE_DIRECTORY, batch_size=BATCH_SIZE)
val_generator = datagen(val_files, IMAGE_DIRECTORY, batch_size=BATCH_SIZE)

history = model.fit(
    train_generator,
    steps_per_epoch=STEPS_PER_EPOCH,
    epochs=EPOCHS,
    validation_data=val_generator,
    validation_steps=VALIDATION_STEPS
)
print("Training complete.")

def plot_loss(history):
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss Over Epochs')
    plt.ylabel('Loss (MAE)')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_loss(history)

def test_and_visualize_with_metrics(model, image_files, image_dir, num_samples=3):
    print("\nVisualizing results with test images and metrics...")

    for file in np.random.choice(image_files, num_samples):
        img_path = os.path.join(image_dir, file)

        original_img = cv2.imread(img_path)
        original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

        img_256 = cv2.resize(original_img, (256, 256))
        img_64 = cv2.resize(img_256, (64, 64))

        input_img = np.expand_dims(img_64 / 255.0, axis=0)
        predicted_img_256 = model.predict(input_img)[0]

        gt_img_tensor = tf.convert_to_tensor(img_256 / 255.0, dtype=tf.float32)
        pred_img_tensor = tf.convert_to_tensor(predicted_img_256, dtype=tf.float32)

        psnr = tf.image.psnr(gt_img_tensor, pred_img_tensor, max_val=1.0).numpy()
        ssim = tf.image.ssim(gt_img_tensor, pred_img_tensor, max_val=1.0).numpy()

        plt.figure(figsize=(12, 4))

        plt.subplot(1, 3, 1)
        plt.title("Low-Res Input (64x64)")
        plt.imshow(img_64)
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.title(f"Model Output (256x256)\nPSNR: {psnr:.2f} dB | SSIM: {ssim:.4f}")
        plt.imshow(predicted_img_256)
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.title("Ground Truth (256x256)")
        plt.imshow(img_256)
        plt.axis('off')

        plt.show()

test_and_visualize_with_metrics(model, val_files, IMAGE_DIRECTORY)
```

Interpretation of the plot : As we can see the graph(both training and validation loss) declines as the number of epochs increased.
